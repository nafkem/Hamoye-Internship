```python
#Normalize the dataset using the MinMaxScaler after removing the following columns: [“date”, “lights”]. 
#The target variable is “Appliances”. Use a 70-30 train-test set split with a random state of 42 (for reproducibility). 
#Run a multiple linear regression using the training set and evaluate your model on the test set. Answer the following questions
#:What is the Mean Absolute Error (in two decimal places)?
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
energy_data1 = energy_data.drop(columns=['date','lights'])
normalised_df = pd.DataFrame(scaler.fit_transform(energy_data1), columns=energy_data1.columns)
X = normalised_df.drop(columns=['Appliances'])
y = normalised_df['Appliances']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
```

```python
#fit the model to the training dataset
linear_model.fit(x_train, y_train)
```

```python
#obtain predictions
predicted_values = linear_model.predict(x_test)
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
print('The Mean Absolute Error (MAE)  is: ',round(mae, 3)) # to 2 decimal place
```

```python
#What is the Residual Sum of Squares (in two decimal places)?

#Residual Sum of Squares (RSS)
import numpy as np
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2) 
```

```python
#What is the Root Mean Squared Error (in three decimal places)?
#Root Mean Square Error (RMSE)
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3) 
```

```python
#What is the Coefficient of Determination (in two decimal places)?
#R-Squared
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2) 
```

```python
#Obtain the feature weights from your linear model above. Which features have the lowest and highest weights respectively?
#comparing the effects of regularisation
def get_weights_df(model, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df

linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')


linear_model_weights.sort_values(by='Linear_Model_Weight', ascending=False)
```
